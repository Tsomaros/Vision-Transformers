```python
import os
import random
import shutil

source_folder = '/content/combined_folder'
destination_folder = '/content/aug_folder'

# Ensure destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Get a list of all .png files
png_files = [f for f in os.listdir(source_folder) if f.lower().endswith('.png')]

# Select random 20%
num_files_to_move = int(len(png_files) * 0.5)
files_to_move = random.sample(png_files, num_files_to_move)

# Move files
for filename in files_to_move:
    shutil.move(os.path.join(source_folder, filename), destination_folder)

print(f"Moved {len(files_to_move)} files to {destination_folder}")
```

```python
# Converts the format of each label in the dataframe from "LabelA|LabelB|LabelC"
# into ["LabelA", "LabelB", "LabelC"], concatenates the
# lists together and removes duplicate labels
unique_labels = np.unique(
    data['Finding Labels'].str.split("|").aggregate(np.concatenate)
).tolist()

print(f"Dataset contains the following labels:\n{unique_labels}")
```

```python
import os
from tqdm import tqdm
import pandas as pd
import numpy as np
from PIL import Image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from concurrent.futures import ProcessPoolExecutor

folder_path = '/content/aug_folder'

png_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.png')]

datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

def process_image(files):
    from PIL import Image
    import numpy as np
    # Load inside the function to avoid issues with multiprocessing
    results = []
    img_path = os.path.join(folder_path, files)
    img = Image.open(img_path).convert('RGB')
    img_array = np.asarray(img)
    img_array = np.expand_dims(img_array, axis=0)
    aug_iter = datagen.flow(img_array, batch_size=1, shuffle=False)

    # You can set how many augmented versions per image you want
    for i in range(1):
        augmented_img = next(aug_iter)[0].astype('uint8')
        new_filename = f"aug_{i}_{files}"
        Image.fromarray(augmented_img).save(os.path.join(folder_path, new_filename))

        # Find the row in the original data
        result = data.loc[data['Image Index'] == files].iloc[0, 1:]
        modified_result = np.insert(result, 0, new_filename)
        results.append(modified_result)

    return results

all_results = []

with ProcessPoolExecutor() as executor:
    for res in tqdm(executor.map(process_image, png_files), total=len(png_files)):
        all_results.extend(res)

# Convert to DataFrame and concat once at the end (much faster!)
modified_df = pd.DataFrame(all_results, columns=data.columns)
data = pd.concat([data, modified_df], ignore_index=True)
data.to_csv('/content/drive/MyDrive/Updated_Data_Entry.csv', index=False)
```

```python
data = pd.read_csv('/content/drive/MyDrive/Updated_Data_Entry.csv')

label_index = {v: i for i, v in enumerate(unique_labels)}

def string_to_N_hot(string: str):
    true_index = [label_index[cl] for cl in string.split("|")]
    label = np.zeros((len(unique_labels),), dtype=float)
    label[true_index] = 1
    return label

data["labels"] = data["Finding Labels"].apply(string_to_N_hot)
```

```python
images_dir = '/content/combined_folder'

output_file_path = os.path.join(images_dir, 'metadata.jsonl')

data[["Image Index", "labels"]].rename(columns={"Image Index": "file_name"}).to_json(output_file_path, orient='records', lines=True)
```
