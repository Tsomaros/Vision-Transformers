```python
import json
import matplotlib.pyplot as plt
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/imagenet-a'

dir_list = os.listdir(path)

markers = ['>','o','v','^']

accuracies = []
parameters = []
model_Id = []

for files in dir_list:
  file = os.path.join(path, files, 'eval_results.json')
  with open(file, 'r') as f:
      data = json.load(f)
      # Extract values and append to lists
      model_Id.append(files)
      try:
        accuracies.append(data['eval_accuracy'])
      except:
        accuracies.append(data['eval_validation_accuracy'])
      parameters.append(data['parameters'])

plt.figure(figsize=(10, 6))
for i in range(len(model_Id)):
  plt.annotate(model_Id[i], xytext = (parameters[i] + 1, accuracies[i]), xy = (parameters[i], accuracies[i]))
  plt.scatter(parameters[i], accuracies[i], label=model_Id[i], marker=markers[i % len(markers)])

plt.xlabel('Parameters')
plt.ylabel('Eval Accuracy')

# Place legend outside of the plot
plt.legend(title='Model Id', loc="center left", bbox_to_anchor=(1, 0.5))
plt.grid(False)

# Adjust the layout to fit the legend outside the plot
plt.tight_layout(rect=[0, 0, 0.85, 1])

# Show the plot
plt.show()
```

```python
path = '/content/drive/MyDrive/imagenet-a'

dir_list = os.listdir(path)

markers = ['>','o','v','^']

accuracies = []
GMacs = []
model_Id = []

for files in dir_list:
  file = os.path.join(path, files, 'eval_results.json')
  with open(file, 'r') as f:
      data = json.load(f)
      # Extract values and append to lists
      model_Id.append(files)
      try:
        accuracies.append(data['eval_accuracy'])
      except:
        accuracies.append(data['eval_accuracy'])
      GMacs.append(data['GMacs'])

plt.figure(figsize=(10, 6))
for i in range(len(model_Id)):
  plt.annotate(model_Id[i], xytext = (GMacs[i] + 0.1, accuracies[i]), xy = (GMacs[i], accuracies[i]))
  plt.scatter(GMacs[i], accuracies[i], label=model_Id[i], marker=markers[i % len(markers)])

plt.xlabel('GFlops')
plt.ylabel('Eval Accuracy')

# Place the legend outside of the plot on the right
plt.legend(title='Model Id', loc="center left", bbox_to_anchor=(1, 0.5))
plt.grid(False)

# Adjust layout to fit the legend outside the plot area
plt.tight_layout(rect=[0, 0, 0.85, 1])

# Show the plot
plt.show()
```

```python
# Path to your directory
path = '/content/drive/MyDrive/imagenet-a'

dir_list = os.listdir(path)

# Initialize lists
accuracies = []
parameters = []
model_Id = []

for files in dir_list:
    file = os.path.join(path, files, 'eval_results.json')
    with open(file, 'r') as f:
        data = json.load(f)
        # Extract values and append to lists
        model_Id.append(files)
        try:
            accuracies.append(data['eval_accuracy'])
        except KeyError:
            accuracies.append(data['eval_validation_accuracy'])
        parameters.append(data['parameters'])

# Plotting the bar chart
plt.figure(figsize=(12, 6))

# Creating the bar chart with accuracies as heights
plt.bar(model_Id, accuracies, color='lightblue', edgecolor='black')

# Adding titles and labels
plt.title('Model Accuracy Comparison', fontsize=16)
plt.xlabel('Model ID', fontsize=14)
plt.ylabel('Evaluation Accuracy', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adding grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
