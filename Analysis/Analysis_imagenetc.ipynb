```python
path = '/content/drive/MyDrive/imagenet-c'

distortion = 'fog' #@param ["zoom_blur", "motion_blur", "pixelate", "jpeg_compression", "shot_noise", "gaussian_noise", "brightness", "fog"]

if distortion == 'zoom_blur':
  path = os.path.join(path, 'Blur', 'zoom_blur')
elif distortion == 'motion_blur':
  path = os.path.join(path, 'Blur', 'motion_blur')
elif distortion == 'pixelate':
  path = os.path.join(path, 'Digital', 'pixelate')
elif distortion == 'jpeg_compression':
  path = os.path.join(path, 'Digital', 'jpeg_compression')
elif distortion == 'gaussian_noise':
  path = os.path.join(path, 'Noise', 'gaussian_noise')
elif distortion == 'shot_noise':
  path = os.path.join(path, 'Noise', 'shot_noise')
elif distortion == 'fog':
  path = os.path.join(path, 'Weather', 'fog')
elif distortion == 'brightness':
  path = os.path.join(path, 'Weather', 'brightness')


dir_list = os.listdir(path)

accuracies = []
parameters = []
model_Id = []
temp = []

for files in dir_list:
  file = os.path.join(path, files)
  json_files = [pos_json for pos_json in os.listdir(file) if pos_json.endswith('.json')]
  for json_file in json_files:
    with open(os.path.join(file, json_file), 'r') as f:
      data = json.load(f)
      accuracies.append(data['eval_validation_accuracy'])
      #parameters.append(data['parameters'])
      model_Id.append(data['Model Id'])
```

```python
plt.figure(figsize=(10, 10))

plt.plot(range(1, 6), accuracies[0:5], marker='o', linestyle='--', color='green', label=model_Id[0], linewidth=1)
plt.plot(range(1, 6), accuracies[5:10], marker='o', linestyle='--', color='blue', label=model_Id[5], linewidth=1)
plt.plot(range(1, 6), accuracies[10:15], marker='o', linestyle='--', color='red', label=model_Id[10], linewidth=1)

plt.title(distortion)
plt.xlabel('Severity')
plt.ylabel('Eval Accuracy')
plt.title('Model Comparison: Eval Accuracy Across'+' '+ distortion +' '+'Severity Levels')
plt.legend(title='Model Id', loc= "best")
plt.grid(True)

# Show the plot
plt.show()
```
